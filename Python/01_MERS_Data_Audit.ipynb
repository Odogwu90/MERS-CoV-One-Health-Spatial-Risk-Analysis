Potential One Health Interpretation

Saudi Arabia is the global epicentre of MERS cases and a major hub
for camel trade. The country also hosts the annual Hajj pilgrimage,
which attracts millions of international visitors and creates a
potential pathway for worldwide dissemination of infectious diseases.

# One Health Story Framework

Dromedary camels are the primary animal reservoir of MERS-CoV.

Human infection occurs through direct or indirect contact with infected camels.

Most reported human cases have occurred in Saudi Arabia and neighbouring countries.

Saudi Arabia is also a major hub for camel trade and hosts the annual Hajj pilgrimage.

Millions of pilgrims travel to and from Saudi Arabia every year, creating worldwide human mobility connections.

Although sustained global transmission has not occurred, international travel has resulted in imported MERS cases in several countries.

This demonstrates how animal health, human health, trade and travel are interconnected within a One Health framework.
